```python
import pandas as pd

euv = pd.read_csv("lyman_alpha.csv")
```

```python
euv
```

```python
from datetime import datetime, timedelta
start_date = datetime(1947 ,1,1)
euv['date'] = euv['time (days since 1947-01-01T12:00)'].apply(lambda x: start_date + timedelta(days=x))
```

```python
euv
```

```python
euv = euv.drop(columns=['time (days since 1947-01-01T12:00)'])
```

```python
years_drop = range(1947,1954)
euv['date'] = pd.to_datetime(euv['date'])
```

```python
euv['Year']= euv['date'].dt.year
euv = euv[ ~euv['Year'].isin(years_drop)]
euv['Month']= euv['date'].dt.month
```

```python
print(euv[euv['Year']==1954])   
```

```python
euv = euv.drop(columns=['date'])
euv = euv[['Year', 'Month', 'irradiance (W/m^2)','uncertainty (W/m^2)']]
```

```python
# Group by 'Year' and 'Month' and calculate the mean for other columns
euv = euv.groupby(['Year', 'Month']).mean().reset_index()

# Display the updated DataFrame
print(euv)
```

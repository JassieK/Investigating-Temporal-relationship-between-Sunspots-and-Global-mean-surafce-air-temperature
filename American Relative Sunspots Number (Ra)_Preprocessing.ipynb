```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
sunspot = pd.read_csv("monthly.csv")
```

```python
sunspot
```

```python
# 6. Yearly Average Sunspots (Trend Analysis)
yearly_avg = sunspot.groupby('Year')['Ra'].mean()
# Font settings
label_fontsize = 14
title_fontsize = 16
axis_fontweight = 'bold'
ticks_fontsize = 12  # Font size for x and y axis ticks
plt.figure(figsize=(12, 6))
sns.lineplot(x=yearly_avg.index, y=yearly_avg.values, marker="o", color='black')
plt.title("Average Sunspot Numbers by Year", fontsize=title_fontsize, fontweight=axis_fontweight)
plt.xlabel("Year",fontsize=label_fontsize, fontweight=axis_fontweight)
plt.ylabel("Average Sunspot Number (Ra)",fontsize=label_fontsize, fontweight=axis_fontweight)
plt.xticks(fontsize=ticks_fontsize, fontweight=axis_fontweight)
plt.yticks(fontsize=ticks_fontsize, fontweight=axis_fontweight)

plt.grid(True)
plt.show()
```

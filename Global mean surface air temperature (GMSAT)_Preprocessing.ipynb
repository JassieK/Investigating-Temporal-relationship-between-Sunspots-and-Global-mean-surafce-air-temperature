```python
Atm = Atm.drop(columns=['J-D', 'D-N', 'DJF', 'MAM', 'JJA', 'SON'])
```

```python
# Visualize the yearly data
plt.figure(figsize=(12, 8))
sns.lineplot(data=Atm.set_index('Year').dropna(), markers=True)
plt.title('Global Temperature Anomalies (1880-Present)')
plt.ylabel('Temperature Anomaly (Â°C)')
plt.xlabel('Year')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1), title="Months")
plt.grid(True)
plt.show()
```

```python
# Unpivot the DataFrame
df_melted = pd.melt(Atm, id_vars=['Year'], var_name='Month', value_name='Value')

# Display the transformed DataFrame
print(df_melted)
```

```python
# Sort the data by the 'Year' column in ascending order
sorted_data = df_melted.sort_values(by='Year', ascending=True)
```

```python
sorted_data
```

```python
# Define custom sorting order for months
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Convert the 'Month' column to a categorical type with the defined order
df_melted['Month'] = pd.Categorical(df_melted['Month'], categories=month_order, ordered=True)

# Sort the data by 'Year' and then by 'Month'
sorted_data = df_melted.sort_values(by=['Year', 'Month'])
```

```python
sorted_data
```

```python
sorted_data.isnull().sum()
```

```python
years_drop = range(1880,1954)
sorted_data = sorted_data[ ~sorted_data['Year'].isin(years_drop)]
```
